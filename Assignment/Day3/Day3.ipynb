## How strongly length of service correlated with starting age?

```python
Starting_Age = df4['Starting Age']
```

```python
Starting_Age
```

```python
Length_of_Service = df4['Length of Service (years)']
```

```python
Length_of_Service
```

### Covariance -

```python
print('Covariance = ', np.cov(Starting_Age, Length_of_Service))
```

### Correlation  -

```python
print('Correlation = ', stats.pearsonr(Starting_Age, Length_of_Service))
```

```python
Starting_Age = df4['Starting Age'].tolist()
Length_of_Service = df4['Length of Service (years)'].tolist()

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.scatter(Starting_Age, Length_of_Service, color = 'r')
ax.set_xlabel('Starting_Age')
ax.set_ylabel('Length_of_Service')
ax.set_title(' : Relationship Between Starting_Age and Length_of_Service : ')
```
